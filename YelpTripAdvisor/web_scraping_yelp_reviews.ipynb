```python

from requests import get
import pandas as pd

url = 'https://www.yelp.com/biz/anchor-bar-buffalo?frvs=True&osq=Restaurants'

response = get(url)
from bs4 import BeautifulSoup

html_soup = BeautifulSoup(response.text, 'html.parser')

review_containers = html_soup.find_all('div', class_ = 'review-content')

reviews = []
scores = []

for container in review_containers:

        review = container.find('p',lang="en").text
        reviews.append(review)

        score = float(container.find('div', class_ = 'biz-rating biz-rating-large clearfix').img.attrs['alt'][0])
        scores.append(score)

first_page_df = pd.DataFrame({'Reviews': reviews,'Score': scores,})
first_page_df
```

```python
first_page_df.to_csv('first_page_review.csv')
```

```python
headers = {"Accept-Language": "en-US, en;q=0.5"}
pages = [str(i) for i in range(20,920,20)]

from time import sleep
from random import randint
from requests import get
import pandas as pd
from time import time
from IPython.core.display import clear_output
from bs4 import BeautifulSoup
from warnings import warn

reviews = []
scores= []

start_time = time()
requests = 0

for page in pages:

        response = get('https://www.yelp.com/biz/anchor-bar-buffalo?start=' + page, headers = headers)

        # Pause the loop
        sleep(randint(8,15))

        # Monitor the requests
        requests += 1
        elapsed_time = time() - start_time
        print('Request:{}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))
        clear_output(wait = True)
        #A successful request is indicated by a status code of 200
        # Throw a warning for non-200 status codes
        if response.status_code != 200:
            warn('Request: {}; Status code: {}'.format(requests, response.status_code))

        # Break the loop if the number of requests is greater than expected
        if requests > 72:
            warn('Number of requests was greater than expected.')  
            break 

        # Parse the content of the request with BeautifulSoup
        page_html = BeautifulSoup(response.text, 'html.parser')

        
        rv_containers = page_html.find_all('div', class_ = 'review-content')

        # For every review
        for container in rv_containers:
           

                        # The review
                review = container.find('p',lang="en").text
                reviews.append(review)

                # The score
                score = float(container.find('div', class_ = 'biz-rating biz-rating-large clearfix').img.attrs['alt'][0])
                scores.append(score)
```

```python
restaurant_ratings = pd.DataFrame({'Reviews': reviews,
                              'Score': scores,
                              })
restaurant_ratings.head(10)
```

```python
restaurant_ratings.to_csv('Anchor_bar_restaurant_reviews.csv')# About 1000 comments and ratings included
```
