```python
import collections
import pandas as pd
words=[]
total_w_c=[]
word=[]
counts=[]
file=open("comment.txt","r+")
wordcount={}
for word in file.read().split():
    
        word = word.lower()

        if word not in wordcount:
            wordcount[word] = 1
        else:
            wordcount[word] += 1
for count,v in wordcount.items():
    words.append(count)
    total_w_c.append(v)
    
test_df = pd.DataFrame({'Words':words,'Counts': total_w_c})

df=test_df.sort_values(by=['Counts'],ascending=False)
df.reset_index(drop='index', inplace=True)
df=df[['Words','Counts']]
df
```

```python
df.to_csv('word_count.csv')
```
