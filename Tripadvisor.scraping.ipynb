```python
from requests import get
import pandas as pd

url = 'https://www.tripadvisor.com/Hotel_Review-g60974-d8518794-Reviews-or25-Buffalo_Marriott_HARBORCENTER-Buffalo_New_York.html#REVIEWS'
response = get(url)
from bs4 import BeautifulSoup

html_soup = BeautifulSoup(response.text, 'html.parser')

review_containers = html_soup.find_all('q', class_ = "hotels-hotel-review-community-content-review-list-parts-ExpandableReview__reviewText--1OjOL")

reviews = []

#print(review_containers)
for container in review_containers:

        review = container.text
        reviews.append(review)

        

first_page_df = pd.DataFrame({'Review': reviews})
first_page_df
```

```python
headers = {"Accept-Language": "en-US, en;q=0.5"}
pages = [str(i) for i in range(5,185,5)]

from time import sleep
from random import randint
from requests import get
import pandas as pd
from time import time
from IPython.core.display import clear_output
from bs4 import BeautifulSoup
from warnings import warn

reviews = []
scores= []
dates= []

start_time = time()
requests = 0

for page in pages:
#url = 'https://www.tripadvisor.com/Hotel_Review-g60974-d8518794-Reviews-or25-Buffalo_Marriott_HARBORCENTER-Buffalo_New_York.html#REVIEWS'

        response = get('https://www.tripadvisor.com/Hotel_Review-g60974-d8518794-Reviews-or' +page+'-Buffalo_Marriott_HARBORCENTER-Buffalo_New_York.html#REVIEWS', headers = headers)

        # Pause the loop
        sleep(randint(12,15))

        # Monitor the requests
        requests += 1
        elapsed_time = time() - start_time
        print('Request:{}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))
        clear_output(wait = True)
        #A successful request is indicated by a status code of 200
        # Throw a warning for non-200 status codes
        if response.status_code != 200:
            warn('Request: {}; Status code: {}'.format(requests, response.status_code))

        # Break the loop if the number of requests is greater than expected
        if requests > 72:
            warn('Number of requests was greater than expected.')
            break

        # Parse the content of the request with BeautifulSoup
        page_html = BeautifulSoup(response.text, 'html.parser')


        review_containers = page_html.find_all('q', class_ = "hotels-hotel-review-community-content-review-list-parts-ExpandableReview__reviewText--1OjOL")



        #print(review_containers)
        for container in review_containers:

            review = container.text
            #print(review)
            reviews.append(review)

        

review_df = pd.DataFrame({'Review': reviews})
review_df
```

```python
frames = [first_page_df, review_df]
result = pd.concat(frames)
result.to_csv('hotel_reviews.csv')
```
