```python
import collections
import pandas as pd
words=[]
total_w_c=[]
word=[]
w=[]
key_words1=['spicy','very','like','wings','good','test','nice','excellent','people','food','really','go','eat']
file=open("comment.txt","r+")
wordcount={}
for word in file.read().split():
    
        word = word.lower()
        for i in key_words1:
            if i==word:
                if word not in wordcount:
                    wordcount[word] = 1
                else:
                    wordcount[word] += 1
for w,v in wordcount.items():
    words.append(w)
    total_w_c.append(v)
    
test_df = pd.DataFrame({'Words':words,'Counts': total_w_c})

df=test_df.sort_values(by=['Counts'],ascending=False)
df.reset_index(drop='index', inplace=True)
df=df[['Words','Counts']]
df
```

```python
df.to_csv('Reviews_specific_words.csv')
```
